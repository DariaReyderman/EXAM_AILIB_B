### Cleaning Duplicates

```python
items = pd.read_csv("items.csv")
items_df = items.copy()

items_dup = items_df[items_df['item_name'].duplicated(keep=False)].sort_values('item_name')
display(items_dup)

items_dup['not_nan_count'] = items_dup.notna().sum(axis=1)
items_dup.sort_values(['item_name', 'not_nan_count'], ascending=[True, True], inplace=True)
display(items_dup)

to_remove = items_dup.drop_duplicates(subset='item_name', keep='first')
to_keep = items_dup[~items_dup['id'].isin(to_keep['id'])]
display('to_keep', to_keep)
display('to_remove', to_remove)
```

```python
order_item = pd.read_csv("order_item.csv")
order_item_df = order_item.copy()

order_item_df.sort_values('item_id', inplace=True)
display(order_item_df)

merged = to_remove.merge(to_keep, how='inner', on='item_name', suffixes=('_removed', '_kept'))
display(merged)

dict_to_replace = dict(zip(merged['id_removed'], merged['id_kept']))
    
order_item_df['item_id'] = order_item_df['item_id'].replace(dict_to_replace)

order_item_df.head(50).sort_values(by='id')
```

```python
customers_df = customers.copy()
order_item_df = order_item.copy()
orders_df = orders.copy()

cleaned_customers = customers_df[customers_df.isnull().sum(axis=1) < 3]
display('cleaned_customers', cleaned_customers)

merged = pd.merge(order_item_df, orders_df[['order_id', 'customer_id']], on='order_id', how='left')
display('merged', merged)

filtered = merged[merged['customer_id'].isin(cleaned_customers['id'])]
display('filtered', filtered)
```

```python
order_item_df = order_item.copy()
items_df = items.copy()

items_df = items_df.rename(columns={'id': 'item_id'})
merged = pd.merge(order_item_df, items_df[['item_id', 'item_price']], on='item_id', how='left')
merged['total_price'] = merged['item_price'] * merged['quantity']
display(merged.head())

total = merged.groupby('order_id')['total_price'].sum().reset_index()
display(total.tail())

statistics = merged.groupby('order_id')['total_price'].agg(['min', 'max', 'mean']).reset_index()
display(statistics)
```

```python
df['year'] = pd.to_datetime(df['joining_date'], errors='coerce').dt.year

sns.countplot(data=df, x='year', palette='viridis')
plt.title('New clients per year')
plt.xlabel('Year'), plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Exercise 6

```python
df_o = pd.read_csv("orders.csv")
df_c = pd.read_csv("customers.csv")

df_c = df_c.rename(columns={'id': 'customer_id'})
df_c['age'] = pd.to_numeric(df_c['age'], errors='coerce')
merged = df_o.merge(df_c[['customer_id', 'age']], on='customer_id', how='left')

sns.histplot(merged, x='age', bins=20, kde=False, color='mediumseagreen')
plt.title('Orders number per age')
plt.xlabel('Age'), plt.ylabel('Orders')
plt.tight_layout()
plt.show()

# There is more purchasing in 40-50 age group and less purchasing in 0-20 and 60-80 age groups
```
